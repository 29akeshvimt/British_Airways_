```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
```

```python
pages = 38
page_size = 100
reviews = []
# for i in range(1, pages + 1):
for i in range(1, pages + 1):
  print(f"Scraping page {i}")
  # Create URL to collect links from paginated data
  url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"
  # Collect HTML data from this page\n"
  response = requests.get(url)
  # Parse content\n"
  content = response.content
  parsed_content = BeautifulSoup(content, 'html.parser')
  for para in parsed_content.find_all("div", {"class": "text_content"}):
   reviews.append(para.get_text())
  print(f"   ---> {len(reviews)} total reviews")
```

```python


import pandas as pd
import re


file_path = '/content/British_Airways_reviews.csv'
df = pd.read_csv(file_path)

# Define the text cleaning function
def clean_text(text):
    # Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply the cleaning function to the 'reviews' column
df['reviews'] = df['reviews'].apply(clean_text)

# Remove rows with null values in the 'reviews' column
df.dropna(subset=['reviews'], inplace=True)

# Save the cleaned DataFrame to a new CSV file
cleaned_file_path = '/content/British_Airways_reviews_cleaned.csv'
df.to_csv(cleaned_file_path, index=False)

print("Data cleaning completed successfully.")
```

```python
df = pd.read_csv("British_Airways_reviews.csv")
df
```

```python
def analyze_sentiment(review_text):
    # Create a TextBlob object
    blob = TextBlob(review_text)
    # Calculate the polarity (-1 to 1 scale, where -1 is negative, 0 is neutral, and 1 is positive)
    sentiment = blob.sentiment.polarity
    return sentiment
```

```python
df['sentiment'] = df['reviews'].apply(analyze_sentiment)
```

```python
df['sentiment_label'] = df['sentiment'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))
```

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove non-alphabetic characters
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)
```

```python
df['cleaned_reviews'] = df['reviews'].apply(preprocess_text)
```

```python
all_reviews = ' '.join(df['cleaned_reviews'])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_reviews)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Save the word cloud image
wordcloud.to_file('/content/British_Airways_wordcloud.png')

print("Word cloud created successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the DataFrame with sentiment scores
df = pd.read_csv("/content/British_Airways_reviews_updated.csv")

# Set up the aesthetics for the plots
sns.set(style="whitegrid")

# 1. Histogram of Sentiment Scores
plt.figure(figsize=(12, 6))
sns.histplot(df['sentiment'], bins=30, kde=True, color='blue')
plt.title('Histogram of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()

# 2. Box Plot of Sentiment Scores
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['sentiment'], color='orange')
plt.title('Box Plot of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.show()

# 3. Density Plot of Sentiment Scores
plt.figure(figsize=(12, 6))
sns.kdeplot(df['sentiment'], shade=True, color='green')
plt.title('Density Plot of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Density')
plt.show()
```

```python

# 4. Sentiment by Review Length
df['review_length'] = df['reviews'].apply(len)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='review_length', y='sentiment', data=df, color='red')
plt.title('Sentiment vs. Review Length')
plt.xlabel('Review Length')
plt.ylabel('Sentiment Score')
plt.show()
```
